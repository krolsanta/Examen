```python
#Ejercicios con bucle while 5 de 5
#El programa pide al usuario que introduzca un número entre 1 y 100.
#El programa entonces genera un número aleatorio entre 1 y 100 y le dice al usuario si su número es mayor,
#menor o igual al número aleatorio.
#El programa continúa hasta que el usuario adivina el número aleatorio.

#Ejercio #1.(while)Adivinar un número:
import random  # Importa el módulo random para generar números aleatorios
# Genera un número aleatorio entre 1 y 100
numero_aleatorio = random.randint(1,100)
numero_usuario = 0  # Inicializa el número del usuario en 0
# Comienza un bucle mientras el número del usuario no sea igual al número aleatorio
while numero_usuario != numero_aleatorio:
    # Pide al usuario que introduzca un número entre 1 y 100
    numero_usuario = int(input("Introduce un número entre 1 y 100: "))
    # Comprueba si el número del usuario es mayor que el número aleatorio
    if numero_usuario > numero_aleatorio:
        print("El número es menor.")
    # Comprueba si el número del usuario es menor que el número aleatorio
    elif numero_usuario < numero_aleatorio:
        print("El número es mayor.")
# Si el bucle se termina, significa que el usuario adivinó el número
print("¡Felicidades! Has adivinado el número.")

#Ejecio #2.(while)Calcular la suma de los números pares del 1 al 100:
#El programa calcula la suma de todos los números pares entre 1 y 100.
suma = 0
numero = 2  # Empezamos en 2, ya que es el primer número par
while numero <= 100:
    suma += numero
    numero += 2  # Incrementamos en 2 para pasar al siguiente número par
print("La suma de los números pares del 1 al 100 es:", suma)

#Ejercio #3.(while)Contar las vocales en una cadena:
#El programa pide al usuario que introduzca una cadena de texto y luego cuenta el número de vocales en la cadena.
# Pide al usuario que introduzca una cadena de texto
cadena = input("Introduce una cadena de texto: ")
# Inicializa un contador para las vocales
contador_vocales = 0
# Define una lista de vocales
vocales = ['a', 'e', 'i', 'o', 'u']
# Itera sobre cada carácter en la cadena de texto
for caracter in cadena:
    # Si el carácter está en la lista de vocales, aumenta el contador
    if caracter.lower() in vocales:
        contador_vocales += 1
# Imprime el resultado
print("El número de vocales en la cadena es:", contador_vocales)

#Ejecio #4.(while)Imprimir los números impares del 1 al 50:
# Itera sobre los números del 1 al 50
for numero in range(1, 51):
    # Verifica si el número es impar
    if numero % 2 != 0:
        # Imprime el número si es impar
        print(numero)

#Ejercio #5.(while) Factorial de un número:
#El programa pide al usuario que introduzca un número y luego calcula el factorial de ese número.
# Función para calcular el factorial de un número
def factorial(n):
    # El factorial de 0 es 1
    if n == 0:
        return 1
    # Calcula el factorial recursivamente
    return n * factorial(n - 1)
# Pide al usuario que introduzca un número
numero = int(input("Introduce un número para calcular su factorial: "))
# Verifica si el número es negativo
if numero < 0:
    print("No se puede calcular el factorial de un número negativo.")
else:
    # Calcula el factorial del número introducido por el usuario
    resultado = factorial(numero)
    # Imprime el resultado
    print("El factorial de", numero, "es:", resultado)
```

```python
#Ejercicios con bucles for 5 de 5
#Ejercio #1.(for) Imprimir los números pares del 1 al 100:
# Itera sobre los números del 1 al 100
for numero in range(1, 101):
    # Verifica si el número es par
    if numero % 2 == 0:
        # Imprime el número si es par
        print(numero)

#Ejercio #2.(for) Calcular la suma de los cuadrados de los números del 1 al 10:
# Inicializa una variable para almacenar la suma de los cuadrados
suma_cuadrados = 0
# Itera sobre los números del 1 al 10
for numero in range(1, 11):
    # Calcula el cuadrado del número y añádelo a la suma
    suma_cuadrados += numero ** 2
# Imprime la suma de los cuadrados
print("La suma de los cuadrados de los números del 1 al 10 es:", suma_cuadrados)

#Ejercio #3.(for) Imprimir una tabla de multiplicar:
# Número para el cual imprimir la tabla de multiplicar
numero = 5
# Imprime el encabezado de la tabla
print("Tabla de multiplicar del", numero, ":")
# Itera sobre los números del 1 al 10
for i in range(1, 11):
    # Calcula el producto del número y el índice actual
    producto = numero * i
    # Imprime la multiplicación
    print(numero, "x", i, "=", producto)

#Ejercio #4.(for) Factorial de un número:
# Función para calcular el factorial de un número
def factorial(n):
    # Inicializa el factorial como 1
    factorial_resultado = 1
    # Itera desde 1 hasta n
    for i in range(1, n + 1):
        # Multiplica el factorial actual por el número actual del bucle
        factorial_resultado *= i
    # Devuelve el resultado
    return factorial_resultado
# Número para calcular el factorial
numero = int(input("Introduce un número para calcular su factorial: "))
# Calcula y muestra el factorial del número dado
print("El factorial de", numero, "es:", factorial(numero))

#Ejercio #5.(for) Contar las letras en una cadena:
# Función para contar las letras en una cadena
def contar_letras(cadena):
    # Inicializa un diccionario para almacenar el recuento de cada letra
    recuento_letras = {}
    # Itera sobre cada letra en la cadena
    for letra in cadena:
        # Si la letra ya está en el diccionario, incrementa su recuento
        if letra in recuento_letras:
            recuento_letras[letra] += 1
        # Si la letra no está en el diccionario, la inicializa con un recuento de 1
        else:
            recuento_letras[letra] = 1
    # Devuelve el diccionario de recuento de letras
    return recuento_letras
# Cadena de ejemplo
cadena_ejemplo = "Hola mundo"
# Llama a la función contar_letras y muestra el resultado
resultado = contar_letras(cadena_ejemplo)
print("Recuento de letras en la cadena:", resultado)
```
